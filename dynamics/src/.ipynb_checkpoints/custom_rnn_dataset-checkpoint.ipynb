```python
import os
import numpy as np
from data_loader import DataLoader
```

```python
root = os.getcwd()
X = np.load(root+"/../datasets/states.npy")
A = np.load(root+"/../datasets/actions.npy")
Y = np.load(root+"/../datasets/next_states.npy")
X_cat = np.concatenate((X, A), axis=2)
```

```python
print(X.shape, A.shape, X_cat.shape, Y.shape)
```

```python
X.shape[0]//100
```

```python
input_states = []
input_actions = []
intermediate_states = []
output_states = []
for n in range(X.shape[0]//100):
    x = X[n]    # shape (T=100, D=21)
    a = A[n]    # shape (T=100, D=8)
    for t in range(1, x.shape[0]):
        input_state_sequence = np.concatenate((x[0:t], np.zeros((x.shape[0]-t, x.shape[1]))), axis=0)
        input_action_sequence = a[:]
        intermediate_state = None
        output_state_sequence = None
        if t == x.shape[0]-1:
            intermediate_state = Y[n, 0, :]
            output_state_sequence = np.concatenate((np.zeros((x.shape[0]-1, x.shape[1])), Y[n, 0:1, :]), axis=0)
        else:
            intermediate_state = x[t]
            output_state_sequence = np.concatenate((np.zeros((t+1, x.shape[1])), x[t+1:]), axis=0)
        input_states.append(input_state_sequence)
        input_actions.append(input_action_sequence)
        intermediate_states.append(intermediate_state)
        output_states.append(output_state_sequence)
input_states = np.array(input_states)
input_actions = np.array(input_actions)
intermediate_states = np.array(intermediate_states)
# intermediate_states = np.reshape(intermediate_states, (intermediate_states.shape[0], intermediate_states.shape[1], 1))
output_states = np.array(output_states)
```

```python
def find_split_points(X, max_timesteps=100):
    assert X.shape[1] == max_timesteps
    assert X.shape[-1] == 21
    split_points = []
    key = np.zeros((X.shape[-1], ))
    for n in range(X.shape[0]):
        for t in range(X.shape[1]):
            if (X[n, t+1, :] == key).all():
                split_points.append(t)
                break
            elif not (X[n, X.shape[1]-1-t, :] == key).all():
                split_points.append(X.shape[1]-1-t)
                break
    return split_points
```

```python
import time
a = time.time()
split_points = find_split_points(input_states)
b = time.time()
print(b-a)
```

```python
def get_split_indices(N, train=0.9, val=0.05, test=0.05):
    n = np.arange(N)
    np.random.shuffle(n)
    print(n)
    a = int(N*train)
    b = int(N*val)
    c = int(N*test)
    train = n[0:a]
    val = n[a:a+b]
    test = n[a+b:]
    return train, val, test
```

```python
train_i, val_i, test_i = get_split_indices(N=input_states.shape[0])
```
