```python
import numpy as np
import os
from variational_naive_rnn import SequenceModel, Trainer
import torch
from utils.visualize import *
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python
device 
```

```python
root = os.getcwd()
X_train, A_train, Y_train = np.load(root+"/../datasets/train/states.npy"), np.load(root+"/../datasets/train/actions.npy"), np.load(root+"/../datasets/train/next_states.npy")
X_val, A_val, Y_val = np.load(root+"/../datasets/val/states.npy"), np.load(root+"/../datasets/val/actions.npy"), np.load(root+"/../datasets/val/next_states.npy")
X_test, A_test, Y_test = np.load(root+"/../datasets/test/states.npy"), np.load(root+"/../datasets/test/actions.npy"), np.load(root+"/../datasets/test/next_states.npy")
```

```python
X_train.shape, A_train.shape, Y_train.shape
```

```python
X_val.shape, A_val.shape, Y_val.shape
```

```python
X_test.shape, A_test.shape, Y_test.shape
```

```python
seq_model = SequenceModel(state_dim=21, action_dim=8, latent_dim=256, num_recurrent_layers=1, device=device).to(device)
```

```python
trainer = Trainer(model=seq_model, device=device, learing_rate=1e-5)
```

```python
trainer.load_checkpoint(path=root+"/../models/vanilla_rnn.pth")
```

```python
trainer.train(train=(X_train, A_train, Y_train), val=(X_val, A_val, Y_val), epochs=1000, batch_size=200, shuffle=True)
```

```python
n=10
```

```python
Y_hat = seq_model.predict(x=X_test[n:n+1], a=A_test[n:n+1])
Y = Y_test[n:n+1]
```
